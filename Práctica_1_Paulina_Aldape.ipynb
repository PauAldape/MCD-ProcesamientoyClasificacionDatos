# Data Visualization

### Type Variable

```python
sns.countplot(googleplay['Type'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Type')
```

### Rating of the apps

```python
sns.countplot(googleplay['Content Rating'])
sns.countplot(googleplay['Content Rating']).set_xticklabels(sns.countplot(googleplay['Content Rating']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('Content Rating')
```

### Relation between Type and Content Rating

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Content Rating',hue='Type',data=googleplay)
plt.title('Relation between Type and Content Rating')
plt.show()
```

### Pie-chart for the Type

```python
labels = ['Free', 'Paid', '0']
size = googleplay['Type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 3))
explode = [0, 0.1, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

### Pie-chart for Rating

```python
googleplay['Content Rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```
